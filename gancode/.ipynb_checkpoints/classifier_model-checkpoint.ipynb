## Method #1

Modified from:
https://stackoverflow.com/questions/51803437/how-do-i-use-a-saved-model-in-pytorch-to-predict-the-label-of-a-never-before-see

```python
class Race_CNN(nn.Module): 
    def __init__(self):
        
        super(Race_CNN, self).__init__()
        self.resnet18 = resnet18(pretrained=True)
        num_ftrs = self.resnet18.fc.in_features
        self.resnet18.fc = torch.nn.Linear(num_ftrs, 1)

    def forward(self, x):
        output = self.resnet18(x)
        output = torch.squeeze(output, dim = -1)
        return output
```

```python
model = Race_CNN()
trained_model = torch.load(Path('/users/riya/race/classification_run/best_metric_model.pth'))
model.load_state_dict(trained_model)
```

```python
test_transforms = Compose(
    [
        LoadImage(image_only=True), 
        # hopefully grayscale? I'll use one that is - mainly the .bmp output of segmentation model. 
        # I'll make sure all are grayscale.
        
        # Resize((480,640)),
        AddChannel(),
        ScaleIntensity(),
        ToTensor(),
        Lambda(lambda x: torch.cat([x, x, x], 0)),
        # ToTensor(),
        Normalize(mean=[0.485, 0.456, 0.406],std=[0.229, 0.224, 0.225])
    ]
)
```

### Forming Input Dataframe

```python
classifier_df = pd.DataFrame(columns = ['input', 'type', 'pred']) # didn't keep posterior/interior/etc. information. might want to
```

```python
classifier_df['input'] = os.listdir(classifier_input_path)
```

```python
type_column = [0]*50

for i in range(len(os.listdir(classifier_input_path))):
    if os.listdir(classifier_input_path)[i][-4:] == '.png': # pngs are from GAN, real since paired images
        type_column[i] = 'real'
    elif os.listdir(classifier_input_path)[i][-4:] == '.bmp': # bmp from segmentation model, passed in fake fundus there
            type_column[i] = 'fake'
```

```python
classifier_df['type'] = type_column
```

```python
classifier_df.head()
```

### Performing Predictions

```python
pred_column = [0]*50

for i in range (len(classifier_df['input'])):
    input = test_transforms(classifier_input_path + classifier_df['input'][i])
    input = input.reshape(1, 3, 640, 480) # this is the right shape! double checked 👍
    
    output = model(input)
    pred_column[i] = output.data.item()
```

```python
classifier_df['pred'] = pred_column
```

### Display All 4 Images For Fake Black Segmentations Analysis

**Also be great to consider nonzero pixels. I'll add some code for that. Ooop, I consider that in segmentation_model because all images there (with difference maps and all), not just the special classification cases.**

```python
def get_diff_score(img):
    score_img = np.array(img)
    num_nonzero = np.count_nonzero(score_img)
    diff_score = score_img.sum()/num_nonzero
    
    return diff_score
```

```python
# also copied from segmentation_model

diff_score_arr = []

def generate_difference_maps(arr_ids, path1, path2): # crop is boolean, either crop or no crop. crop if no_crop fails.
    
    size_tuple = (640, 480)
    
    for i in arr_ids:

        seg_img = Image.open(path1 + i + "_real_A" + ".png")
        seg_img = seg_img.resize(size_tuple, Image.BICUBIC).convert('L')
        seg_fake_img = Image.open(path2 + i + "_fake_fundus" + ".bmp") # size (640,480) because of segmentation model, but has been cropped
        
        # print(seg_img.size, seg_fake_img.size)
        # print(seg_img.mode, seg_fake_img.mode)
        
        diff_img = ImageChops.difference(seg_img, seg_fake_img) # absolute value, yay
        
        print(i)
        
        # get diff score
        diff_score = get_diff_score(diff_img)
        print(diff_score)
        diff_score_arr.append(diff_score)
        
        diff_plot = plt.imshow(diff_img, cmap='gray')   
        plt.show(diff_img)
```

```python
generate_difference_maps(fake_black_ids, gan_results_path, segmentation_output_path)
```

```python
np.mean(diff_score_arr)
```

```python
# function copied from segmentation_model

def generate_output_images(arr_ids, path1, path2): # path 1: gan results, path 2: segmentation on fake fundus
    
    size_tuple = (640, 480) # width, height
    
    for i in arr_ids:
        seg_img = Image.open(path1 + i + "_real_A" + ".png").resize(size_tuple, Image.BICUBIC)
        
        # mpimg.imread(path + i + "_real_A" + ".png")
        
        fund_real_img = Image.open(path1 + i + "_real_B" + ".png").resize(size_tuple, Image.BICUBIC)
        # mpimg.imread(path + i + "_real_B" + ".png")
        
        fund_fake_img = Image.open(path1 + i + "_fake_B" + ".png").resize(size_tuple, Image.BICUBIC)
        # fund_fake_img = mpimg.imread(path + i + "_fake_B" + ".png")
        
        seg_fake_img = Image.open(path2 + i + "_fake_fundus" + ".bmp") ## these are bmps! already resized, no need to do that.
        
        # generate 4 x 1 plot
        
        f, axarr = plt.subplots(1,4)
        f.suptitle(i)
        axarr[0].imshow(seg_img)
        axarr[1].imshow(fund_real_img)
        axarr[2].imshow(fund_fake_img)
        axarr[3].imshow(seg_fake_img, cmap='gray')
```

```python
generate_output_images(fake_black_ids, gan_results_path, segmentation_output_path)
```
