## Resizing Black and White Images

pix2pix expects them to be same size (will resize all later to 256 x 256 which stretches them a bit - any issues arise because of this?)

Also, accidently made the original dataset all resized :) so no issue

There are 314 images not of the size (640,480) out of a total 4546 in the fundus images. However, all 4546 segmentation images are the same size: (640,480). Next, I want to make all of these images the same size. I'll resize all images in the new directory that are the wrong size, just by scaling heights and width by same amount. I could also just delete them, but 314 is a significant number of images. 

Conveniently, all of these incorrectly sized images have the right ratio of width:height of 640:480, so resizing won't be too much of a problem!

```python
def wrong_size_check(path, check):
    
    num_wrong_size = 0
    if_cond = 0
    
    for i in os.listdir(path): 
        im = Image.open(path + i)
        
        if check == 'divide':
            if_cond = im.size[1]/im.size[0]
        elif check == 'or':
            if_cond = im.size[1] != 480 or im.size[0] != 640
            
        if (if_cond): # also tested variations of (im.size[1] != 480 or im.size[0] != 640)
            num_wrong_size+=1
            print (i)
    
    return num_wrong_size
```

```python
wrong_size_check(white_testA, 'or'), wrong_size_check(white_testB, 'or')
```

## Converting Black and White Images to .bmp file type

Segmentation images are of ending .bmp vs. fundus images, which are of the ending .png. The .bmp has much more quality (harder to process, but also with more quality). I am thinking to turn .png into .bmp is possible, but may not be.

```python
# use only Btrain_path and Atrain_path!
```

```python
os.listdir(Btrain_path)
```

```python
os.listdir(Btest_path)
```

```python
# Btrain and Btest has pngs. We want to convert them all to bmps.

def convert_to_bmp(path):
    for i in os.listdir(path):
        id = i[:-4]
        Image.open(path + id + ".png").save(path + id + ".bmp")
        os.remove(path + id + ".png")
```

```python
convert_to_bmp(Btrain_path)
convert_to_bmp(Btest_path)
```
